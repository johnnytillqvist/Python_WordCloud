### Word Count

```python
import PyPDF2

# Open the PDF file
pdf_file = open(r'C:\Users\johnn\Desktop\Python\CV.pdf', 'rb')

# Create a PDF reader object
pdf_reader = PyPDF2.PdfReader(pdf_file)

# Count the words in each page
total_words = 0
for page in pdf_reader.pages:
    text = page.extract_text()
    words = text.split()
    total_words += len(words)

# Close the PDF file
pdf_file.close()

# Print the total number of words
print("Total Words:", total_words)
```

### Word Cloud

```python
import PyPDF2
import matplotlib.pyplot as plt
from wordcloud import WordCloud

%matplotlib inline

# Open the PDF file
pdf_file = open(r'C:\Users\johnn\Desktop\Python\CV.pdf', 'rb')

# Create a PDF reader object
pdf_reader = PyPDF2.PdfReader(pdf_file)

# Extract text from each page
text = ''
for page in pdf_reader.pages:
    text += page.extract_text()

# Generate word cloud with custom colors
wordcloud = WordCloud(colormap='Blues').generate(text)  # Change the colormap as per your preference

# Note: https://matplotlib.org/stable/tutorials/colors/colormaps.html

# Configure the plot size and title
plt.figure(figsize=(10, 6))
plt.title('WordCloud')

# Display the generated word cloud using matplotlib
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()

# Close the PDF file
pdf_file.close()
```

### Sentiment Analysis

```python
from textblob import TextBlob

# Open the PDF file
pdf_file = open(r'C:\Users\johnn\Desktop\Python\CV.pdf', 'rb')

# Create a PDF reader object
pdf_reader = PyPDF2.PdfReader(pdf_file)

# Extract text from each page
text = ''
for page in pdf_reader.pages:
    text += page.extract_text()

# Perform sentiment analysis using TextBlob
blob = TextBlob(text)
polarity = blob.sentiment.polarity
subjectivity = blob.sentiment.subjectivity

# Close the PDF file
pdf_file.close()

# Print the sentiment analysis results
print("Sentiment Polarity:", polarity)
print("Sentiment Subjectivity:", subjectivity)
```

- Sentiment polarity: 0.1889 (slightly positive)
- Sentiment subjectivity: 0.2590 (moderately objective)
